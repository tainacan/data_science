```python
#Script para coletar os nomes dos diretórios e arquivos e cruzar com os códigos.
#Tem por objetivo facilitar a montagem do arquivo para upload no Tainacan


import pandas as pd
import os 
```

```python
# base a ser substituídas
base = pd.read_excel('basearquivo.xlsx')
```

```python
base
```

```python
pasta = './Mídias'
for diretorio, subpastas, arquivos in os.walk(pasta):
    for arquivo in arquivos:
        
        #calcula o tamanho do nome do arquivo em caracteres
        tamanhonome=len(arquivo)
        
        #pega apenas o código do arquivo retirando a extensão
        codigoarquivo=arquivo[:tamanhonome-4]
                
        #verifica se possui "P" no nome do arquivo, indicando que possui mais de um arquivo
        x = codigoarquivo.find("P")
        if x!=-1:            
            #isola apenas o código dos nomes de arquivo com P
            #verifica se tem um espaço em branco antes do P
            if codigoarquivo[x-1]==" ":
                codigoarquivo=codigoarquivo[:x-1]
                #print(codigoarquivo)
                #print("Tem ESPAÇO EM BRANCO antes do P!")
            else:
                codigoarquivo=codigoarquivo[:x]
                #print(codigoarquivo)
                #print("Não tem ESPAÇO EM BRANCO antes do P!")
                
        entrou=0
        for i in base.index:
            codigoitem=base['Código do Item'][i] #atentar para nomes das colunas
            codigoitem=codigoitem[11:]
            
            if codigoarquivo==codigoitem:
                entrou=1
                conteudo=str(base['Arquivo principal'][i])
                if len(conteudo)>5:
                    base['Anexos'][i]=os.path.join(os.path.realpath(diretorio), arquivo)
                else:
                    base['Arquivo principal'][i]=os.path.join(os.path.realpath(diretorio), arquivo)
    
        #testa se o código foi identificado e exibe aqueles que não foram encontrados na documentação
        if(entrou==0):
            print("Não achei metadados para :",arquivo)
                    
base.to_excel("basearquivo-EXCEL-NOMESARQUIVOS.xlsx")
        
```
