```python
# base a ser substituídas
base = pd.read_csv('/Users/daltonmartins/Downloads/Museu-do-Colono_tratado.csv')
```

```python
pasta = '/Users/daltonmartins/Downloads/MUSEU DO COLONO'
for diretorio, subpastas, arquivos in os.walk(pasta):

    #mostra o tamanho do vetor
    #print(len(arquivos))
    
    for arquivo in arquivos:
        
        #calcula o tamanho do nome do arquivo em caracteres
        tamanhonome=len(arquivo)
        #pega apenas o código do arquivo retirando a extensão
        codigoarquivo=arquivo[:tamanhonome-4]
        
        #print(arquivo)
        #print(codigoarquivo)
        
        #testa para identificar os separadores "-" e "_" 
        #lembrar de tratar a questão dos espaçamentos em branco
        
        #verifica se existe "-"
        x = codigoarquivo.find("-")
        if x!=-1:
            
            #testa para ver se existe espaço em branco
            y = codigoarquivo.find(" ")
            if y!=-1:
                codigoarquivo=codigoarquivo[:y]
            else:
                codigoarquivo=codigoarquivo[:x]
        else:
            #verifica se existe "-"
            x = codigoarquivo.find("_")
            if x!=-1:
                #testa para ver se existe espaço em branco
                y = codigoarquivo.find(" ")
                if y!=-1:
                    codigoarquivo=codigoarquivo[:y]
                else:
                    codigoarquivo=codigoarquivo[:x]
                    
        #verifica se existe "MC."
        w = codigoarquivo.find("MC.")
        if w!=-1:
            codigoarquivo=codigoarquivo.replace("MC.", "")
                    
        
                
        #print(codigoarquivo)
                
        for i in base.index:
            codigoitem=str(base['Nº de Registro'][i]) #atentar para nomes das colunas

            #Trata codigoitem para substituir "/" por "."
            x=codigoitem.find("/")
            if x!=-1:
                codigoitemnormalizado=codigoitem.replace("/", ".")
            else:
                codigoitemnormalizado=codigoitem
            
             #Trata codigoitem para substituir "/" por "."
            x=codigoitemnormalizado.find(" ")
            if x!=-1:
                codigoitemnormalizado=codigoitemnormalizado.replace(" ", "")     
            
             #Trata codigoitem para substituir "MC." por ""
            x=codigoitemnormalizado.find("MC.")
            if x!=-1:
                codigoitemnormalizado=codigoitemnormalizado.replace("MC.", "")    
            
            if codigoarquivo==codigoitemnormalizado:
                print('.', end=""),
                #na primeira vez que entra, coloca o arquivo como PRINCIPAL
                conteudo=str(base['Arquivo principal'][i])
                if len(conteudo)>5:
                    if len(str(base['Anexos'][i]))>5:
                        base['Anexos'][i]=base['Anexos'][i]+"||"+os.path.join(os.path.realpath(diretorio), arquivo)
                    else:
                        base['Anexos'][i]=os.path.join(os.path.realpath(diretorio), arquivo)
                else:
                    base['Arquivo principal'][i]=os.path.join(os.path.realpath(diretorio), arquivo)
    
        #testa se o código foi identificado e exibe aqueles que não foram encontrados na documentação
        #if(entrou==0):
        #    print("Não achei metadados para :",arquivo)
                    
base.to_excel("baseMUSEUCOLONO-NOMESARQUIVOS.xlsx")
        
```
