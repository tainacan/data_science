```python
#Script para substituir os termos brutos das bases originais pelos termos tratados
#Sempre deve usar a base DE PARA - nomes tratados na primeira coluna e nomes originais na segunda

import pandas as pd
```

```python
# base a ser substituídas
base = pd.read_excel('PalacioCatalogoLevantamento.xlsx')

# base de termos DE PARA
vocabulario = pd.read_excel('vocabnome_tratado_DE_PARA.xlsx')
```

```python
#Tratamento do Levantamento

for i in base.index:
    nomeplanilha=base['autor'][i] #atentar para nomes das colunas
    achou=vocabulario.isin([nomeplanilha])
    total=achou.index.stop
    for j in achou.index:
        if (achou['Nomes originais'][j]):
            
            novostermos=vocabulario['Autor'][j]
            #Trata casos de termos multivalorados na base original e que foram separados em múltiplas linhas
            x=j #pega a posição termo composto na base original
            if x<total-1:
                while ((vocabulario['Nomes originais'][x+1])==''):
                    novostermos=novostermos+'||'+vocabulario['Autor'][x+1]
                    x=x+1
                
            #substitui nas base
            base=base.replace(base['autor'][i],novostermos) #atentar para nomes das colunas

            #imprime na tela para verificação
            print("Nome base:", nomeplanilha)
            print("Nome no vocabulario controlado:",novostermos) #atentar para nomes das colunas
            print("*********************")
            
#atentar para trocar nome do arquivo de saída
base.to_excel("PalacioCatalogoLevantamento-AUTORES-TRATADOS.xlsx")
            
        
            
```
